```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


with open("tot_nreads.txt", "r") as f:
    tot_reads=int(f.read())
dedup_reads=[]
for file in ["nreads_ds2.txt","nreads_ds1.txt","nreads.txt"]:
    with open(file, "r") as f:
        dedup_reads.append(int(f.read()))

dedup05=dedup_reads[1]
dedup03=dedup_reads[0]
dedup1=dedup_reads[2]
nb_reads_out=np.array([dedup03,dedup05,dedup1])
saturation_level=np.array([0.3,0.5,1])
tot_reads=tot_reads*saturation_level
S1=(1-nb_reads_out/tot_reads)[0]
S2=(1-nb_reads_out/tot_reads)[1]
S3=(1-nb_reads_out/tot_reads)[2]
R1=tot_reads[0]
R2=tot_reads[1]
R3=tot_reads[2]
```

```python
from sympy import Matrix, solve_linear_system
from sympy.abc import x, y
system = Matrix(( (R3, -S3, R3*S3), (R2, -S2, R2*S2)))
output=solve_linear_system(system, x, y)
Vmax=float(output[x])
Km=float(output[y])
S_1=np.linspace(0.7,Vmax-2e-3,101)
R_1=S_1*Km/(Vmax-S_1)
plt.plot(R_1/1e6,S_1,label='Dowansampling by 1 and 0.5')
#2nd
system = Matrix(( (R1, -S1, R1*S1), (R2, -S2, R2*S2)))
output=solve_linear_system(system, x, y)
Vmax=float(output[x])
Km=float(output[y])
S_2=np.linspace(0.7,Vmax-4e-3,101)
R_2=S_2*Km/(Vmax-S_2)
plt.plot(R_2/1e6,S_2,label='Dowansampling by 0.5 and 0.3')
#3rd
system = Matrix(( (R3, -S3, R3*S3), (R1, -S1, R1*S1)))
output=solve_linear_system(system, x, y)
Vmax=float(output[x])
Km=float(output[y])
S_3=np.linspace(0.7,Vmax-2e-3,101)
R_3=S_3*Km/(Vmax-S_3)
plt.plot(R_3/1e6,S_3,label='Dowansampling by 1 and 0.3')
plt.scatter(tot_reads/1e6,(1-nb_reads_out/tot_reads),c='r')
plt.legend()
plt.title("Saturation from BAM files")
plt.xlabel("Reads (In Millions)")
plt.ylabel("Saturation")
```
